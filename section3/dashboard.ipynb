```python
import requests
import pandas as pd
# get data from api - https://api.covid19api.com/dayone/country/singapore/status/confirmed
url = 'https://api.covid19api.com/dayone/country/singapore/status/confirmed'
response = requests.get(url,verify=False)
data_json = response.json()
df = pd.DataFrame(data_json)
# df
```

```python
# add year column 
df['year'] = pd.DatetimeIndex(df['Date']).year
# df
```

```python
# add month column
df['month'] = pd.DatetimeIndex(df['Date']).month
# df
```

```python
#combine month year 
df["period"] = df["month"].astype(str) + '-' + df["year"].astype(str)
```

```python
#convert to datetime
df['period'] = pd.to_datetime(df['period'])
# df
```

```python
#get the count per day 
df['cases_per_day'] = df['Cases'].diff()
df
```

```python
import matplotlib.pyplot as plt 
df1 = df.groupby('period', as_index=False)['Cases'].mean()
df1
```

```python
plt.plot(df1['period'], df1['Cases'])
plt.title('Number of cases in Singapore over time')
plt.xlabel('Month-Year')
plt.ylabel('Number of Cases')
plt.xticks(rotation=45, ha='right')
plt.show()
```
