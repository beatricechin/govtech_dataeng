```python
import pandas as pd
df = pd.read_csv("car.data",index_col=False,names=["buying", "maint", "doors", "persons","lug_boot","safety","class"])
df.head(10)
```

```python
#get the variables needed

features = df[['maint','doors','lug_boot','safety','class']]
# df = df[(df['maint'] == 'high') & (df['doors'] == '4') & (df['lug_boot'] == 'big') & (df['safety'] == 'high') &  (df['class'] == 'good')]
features.head(10)
```

```python
# convert to numeric 
from sklearn.preprocessing import LabelEncoder
num = LabelEncoder()
df1 = features[['maint','doors','lug_boot','safety','class']]
df1['maint_num'] = num.fit_transform(features['maint'])
df1['doors_num'] = num.fit_transform(features['doors'])
df1['lugBoot_num'] = num.fit_transform(features['lug_boot'])
df1['safety_num'] = num.fit_transform(features['safety'])
df1['class_num'] = num.fit_transform(features['class'])
df1.tail(10)
```

```python
#keep only the numeric columns
df1 = df1[['maint_num', 'doors_num','lugBoot_num','safety_num','class_num']]
df1.head(10)
```

```python
# also convert the target : buying
df["buying_num"] = num.fit_transform(df["buying"].astype('str'))
df.head(10)
```

```python
#keep only the numeric columns
target = df[['buying_num']]
target.head(10)
```

```python
from sklearn.model_selection import train_test_split

#split data set into train and test sets
data_train, data_test, target_train, target_test = train_test_split(df1,target, test_size = 0.30, random_state = 10)
len(data_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3)
clf = clf.fit(data_train,target_train)
y_pred = clf.predict(data_test)
```

```python
import collections
counter=collections.Counter(y_pred)
print(counter)
```

```python
print("Accuracy:",metrics.accuracy_score(target_test, y_pred))
```
